```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
asset1=pd.read_csv('2330台積電.csv')
asset2=pd.read_csv('2912統一超.csv')
asset3=pd.read_csv('2845遠東銀.csv')
```

```python
#設定投資組合權重與投資總資金
#將歷史資料整理為dataframe
date=asset1.交易[2:13]
s1=asset1.收盤[2:13]
s2=asset2.收盤[2:13]
s3=asset3.收盤[2:13]
#權重
w1=0.5
w2=0.3
w3=0.2
p=w1*s1+w2*s2+w3*s3
#總資金
i=1000000
#分別股數
share1=i/p
f= {
    "date": date,
    "portfolio": p,
    "2330": s1,
    "2912": s2,
    "2845": s3
}
data = pd.DataFrame(f)
print(w1,w2,w3)
data
```

```python
#計算報酬率並整理為dataframe
return_=data
for i in range(2,12):
    return_.portfolio[i]=(return_.portfolio[i+1]-return_.portfolio[i])/return_.portfolio[i]
    return_['2330'][i]=((data['2330'][i]-data['2330'][i+1])/data['2330'][i+1])
    return_['2330'][i]=float(return_['2330'][i])*100
    return_['2912'][i]=(data['2912'][i]-data['2912'][i+1])/data['2912'][i+1]
    return_['2912'][i]=float(return_['2912'][i])*100
    return_['2845'][i]=(data['2845'][i]-data['2845'][i+1])/data['2845'][i+1]
    return_['2845'][i]=float(return_['2845'][i])*100
print('單位:(%)')
return_[0:10]
```

```python
date_list=[]
p1=[]
p2=[]
p3=[]
v=[] #portfolio
for i in range(10):
    #date設定
    k=10+i+1
    k=str(k)
    d='6月'+k+'日'
    date_list.append(d)
    #模擬資產價格
    j=i+2
    a1=s1[2]*(1+return_['2330'][j]*0.01)
    a2=s2[2]*(1+return_['2912'][j]*0.01)
    a3=s3[2]*(1+return_['2845'][j]*0.01)
    p1.append(a1)
    p2.append(a2)
    p3.append(a3)
    v.append(w1*a1+w2*a2+w3*a3)
```

```python
f_pre= {
    "date": date_list,
    "portfolio": v,
    "2330": p1,
    "2912": p2,
    "2845": p3
}
data_pre = pd.DataFrame(f_pre)
data_pre
```

```python
dv=[p[2]]
dp1=[s1[2]]
dp2=[s2[2]]
dp3=[s3[2]]
for i in range(9):
    d=v[i+1]-v[i]
    d1=p1[i+1]-p1[i]
    d2=p2[i+1]-p2[i]
    d3=p3[i+1]-p3[i]
    dv.append(d)
    dp1.append(d1)
    dp2.append(d2)
    dp3.append(d3)
f_diff= {
    "date": date_list,
    "portfolio": dv,
    "2330": dp1,
    "2912": dp2,
    "2845": dp3
}
data_diff = pd.DataFrame(f_diff)
data_diff
```
