```python
s0=float(input("請輸入初始標的資產價格"))
n=int(input("請輸入欲進行的randomwalk模擬次數"))
epislon=np.random.random(n)
asset=[0 for i in range(n)]
asset[0]=s0
for i in range(1,n):
    if epislon[i]>0.5:
        epislon[i]=1
    else:
        epislon[i]=-1
    asset[i]=asset[i-1]+epislon[i]
t=[i for i in range(n)]
plt.plot(t,asset)
```
